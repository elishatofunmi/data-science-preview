```python
import numpy

def deal_with_outliers(data):
    new_dict = {}
    for k in data.columns:
        if isinstance(data[k][2], int):
            mean = numpy.mean(data[k], axis=0)
            sd = numpy.std(data[k], axis=0)

            final_list = [x if (x > mean - 2 * sd) else median for x in list(data[k])]
            final_list = [x if (x < mean + 2 * sd) else median for x in final_list]
            new_dict[k] = final_list
        elif isinstance(data[k][2], float):
            mean = numpy.mean(data[k], axis=0)
            sd = numpy.std(data[k], axis=0)
            median = numpy.median(data[k], axis = 0)

            final_list = [median if (x > mean - 2 * sd) else x for x in list(data[k])]
            final_list = [median if (x < mean + 2 * sd) else x for x in final_list]
            new_dict[k] = final_list
        else:
            new_dict[k] = data[k]
            
    dataframe = pd.DataFrame(new_dict)
    return dataframe

new_dataframe = deal_with_outliers(read_credit)
new_dataframe.head()
```

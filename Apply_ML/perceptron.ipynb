```python
import tensorflow as tf
import os
# disables all tensorflow logs/errors
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 

import warnings
warnings.filterwarnings('ignore')


import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

iris = load_iris()
x = iris.data[:,(2,3)]
y = (iris.target == 0).astype(np.int)

per_clf = Perceptron(random_state = 42)
per_clf.fit(x, y)
y_pred = per_clf.predict([[2,0.5]])
```

```python
feature_cols = ['petal', 'sepal']
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units = [3,5], n_classes = 3)
dnn_clf = tf.contrib.learn.SKCompation(dnn_clf)
dnn_clf.fit(x, y, batch_size = 50, steps = 10)
```

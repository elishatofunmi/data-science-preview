```python
import pandas as pd
import numpy as np
import sklearn as sk
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# A simple machine learning model using the iris dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.DESCR
```

```python
iris.feature_names
```

```python
x, y = iris.data, iris.target
# convert to dataframe
data = pd.DataFrame(x, columns = iris.feature_names)
data['target'] = y
data.head()
```

## Data Visualization

```python
plt.scatter(data['sepal length (cm)'],data['sepal width (cm)'], color = ['r','g'], linestyle = 'dashed', marker = 'o')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width(cm)')
plt.grid('on')
plt.title('plot of sepal length(cm) against sepal width (cm)')
plt.show()
plt.savefig('plot of sepal length(cm) against sepal width (cm)', format = 'png')
```

```python
def plot_data(column_x, column_y):
    plt.scatter(data[column_x],data[column_y], color = ['r','g'], linestyle = 'dashed', marker = 'o')
    plt.xlabel(column_x)
    plt.ylabel(column_y)
    plt.grid('on')
    text = 'plot of ' +column_x + ' against ' + column_y
    plt.title(text)
    plt.show()
    plt.savefig(text, format = 'png')
    return 
```

```python
data.columns
```

```python
plot_data('sepal width (cm)', 'petal length (cm)')
```

```python
plot_data('petal length (cm)', 'petal width (cm)')
```

```python
# boxplot
plt.boxplot(data['petal length (cm)'])
plt.show()
```

```python
def box_plot(x):
    text = 'box plot of column ' +x
    plt.title(text)
    plt.boxplot(data[x])
    plt.grid('on')
    plt.show()
    return
```

```python
box_plot('petal width (cm)')
```
